```python
df=data.copy()
```

```python

df = df[['reviews.title','reviews.text','reviews.date','reviews.rating','categories','brand','name']]
```

```python
df.head(1)
```

```python
import matplotlib.pyplot as plt
```

```python
##how many reviews of each brand
df['brand'].value_counts()
```

```python
df.dtypes
```

```python
df['brand'].value_counts().plot(kind='bar',figsize=(10,5))
plt.title("number of reviews per brand")
plt.xlabel('Brand')
plt.ylabel('number of reviews')
plt.show()
```

```python
##review rating percentage

df['reviews.rating'].value_counts().plot(kind='pie',figsize=(5,5))
plt.title("percentage of review.rating")
plt.show()
```

```python
##hiatograms of review rating
# 
df['reviews.rating'].plot(kind='hist',bins=5)
plt.title("histograms of review rating")
plt.xlabel("Rating")
plt.show()
```

```python
#scatter plot
plt.figure(figsize=(10,5))
plt.scatter(df['reviews.date'],df['reviews.rating'],alpha=0.5)
plt.title("scatter plot of review dDate vs Rating")
plt.xlabel('Review Date')
plt.ylabel('Rating')
plt.show()
```
